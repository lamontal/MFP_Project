```python
Bfast["DoW"] = Bfast["Date"].dt.dayofweek
Lunch["DoW"] = Lunch["Date"].dt.dayofweek
Dinner["DoW"] = Dinner["Date"].dt.dayofweek
Snacks["DoW"] = Snacks["Date"].dt.dayofweek
```

```python
BfastCal = Bfast.groupby("DoW")["Calories"].sum()/(len(Bfast)/7)
LunchCal = Lunch.groupby("DoW")["Calories"].sum()/(len(Lunch)/7)
DinnerCal = Dinner.groupby("DoW")["Calories"].sum()/(len(Dinner)/7)
SnacksCal = Snacks.groupby("DoW")["Calories"].sum()/(len(Snacks)/7)
```

```python
def TDEEcalc(start, end):
    CalLb = 3500
    xscal = (WeightedAvg.loc[start]- WeightedAvg.loc[end]) * CalLb
    TDEE = (rollTC.loc[start:end].sum() - xscal)/len(rollTC.loc[start:end])
    return TDEE
```
